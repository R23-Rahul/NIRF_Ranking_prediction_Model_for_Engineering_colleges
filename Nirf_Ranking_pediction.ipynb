```python
#This lambda function checks if the value x in the 'Rank' column is a digit (numeric) ---> If it is a digit, it keeps the value as it is,
#If it is not a digit--> it removes the last character from the value.
dataset['Rank'] = dataset['Rank'].apply(lambda x: x if str(x).isdigit() else x[:-1])
```

```python
#in data2016, we can see in 'Institute Name' column that the 'of' is written as 'Of'
#and also there is a comma before the name of place (Ex-Indian Institute Of Technology, Madras)
# and in other dataset this shown in differ manner
#so , here we need to convert the data2016 'Insititute Name' as another dataset 
```

```python
#the functions in this module let you check if a particular string matches a given regular expression
#(or if a given regular expression matches a particular string, which comes down to the same thing)
import re
```

```python
#here , the we slected the 'Institute Name' from the data2016 and given its name as 'college'
college = data2016['Institute Name']
#to store the updated name of college
coll = []

for clg in college:
    clg = re.sub(r'Of', 'of', clg) #here , changing the 'Of'-->'of'
    if ',' in clg: # check if there is any comma(,) in 'Institute Name ',if present-->
        clg = ''.join(clg.split(',')) #remove the comma and place '' ,at place of comma
        coll.append(clg) # append the comma romoved name in the coll list
    else:        #if there is no comma(,) in 'Institute Name',then-->
        coll.append(clg)  #appenf the name in the coll list directly
    print(clg)  # Add this line to see the cleaned college names in your output
```

```python
#as the list 'coll' have comma(,)removed and 'Of'-->'of' 'Institute Name', so we update the 'Institute Name' column by 'coll'list
data2016['Institute Name']=coll
```

```python
#now select the features for training the model
selected_features=['TLR', 'RPC', 'GO', 'OI', 'Perception'] # here we select the columns 'TLR', 'RPC', 'GO', 'OI', 'Perception' that will help in prediction of 'Score'
X=dataset[selected_features]
y=dataset['Score']# we select 'Score' as y
```

```python
from sklearn.model_selection import train_test_split #This function is commonly used to split a dataset into training and testing sets for machine learning purposes.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
#here test_size=0.3 shows that  30% of the data will be used as the test set, and the remaining 70% will be used as the training set
# random_state : --> This establishes the random seed to ensure consistency. Your results are repeatable since using a random seed guarantees that the data is divided consistently each time the code is run.
```

```python
from sklearn.linear_model import LinearRegression
# Initialize the Linear Regression model
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train) #fit is used for train the model on linear regression with training data
```

```python
predictions = lr.predict(X_test) #This will use the trained model to predict the target values based on the features in X_test
```

```python
print('R2 Score: ',lr.score(X_test,y_test)) #here we are calucating R square score and printing it
#R2 score tells you how well the linear regression model fits the test data. 
#The score can range from 0 to 1, where 1 indicates a perfect fit,
#and 0 indicates that the model does not explain any variability in the target variable
```

```python
# NOw ,Error Measurment
from sklearn.metrics import mean_squared_error
import math
# Calculate mean squared error to evaluate the model
mse = mean_squared_error(y_test, predictions)
#Mean Squared Error (MSE) is a common metric used to measure the average squared differences between the predicted values and the actual values in a regression problem
# Print the mean squared error
print(f'Mean Squared Error: {mse}')
```

```python
rmse = math.sqrt(mse)
#Mean Squared Error (MSE) is a common metric used to measure the root of average squared differences between the predicted values and the actual values in a regression problem
print(f' Root Mean Squared Error: {rmse}')
```

```python
# as we got the R2 score for Linear Regression about 0.999 , and it is highly accurate although it may be due to over fitting of data
#over fitting --->Overfitting occurs in machine learning when a model learns the training data too well, including its noise and outliers, to the extent that it negatively impacts its performance on new, unseen data.
#so we will check the overfitting is present here or not
```

```python
#The Lasso regression is a type of linear regression that uses L1 regularization,
#which adds a penalty term to the linear regression loss function.
#This penalty term discourages the use of unnecessary features by adding the absolute values of the coefficients of the features to the loss function
from sklearn.linear_model import Lasso
```

```python
lr1=Lasso().fit(X_train,y_train)
lr1.score(X_test,y_test)
```

```python
#Ridge or L2 Regularization
# Ridge regression, another type of linear regression that applies L2 regularization
#This regularization technique discourages large coefficients and can help prevent overfitting.
from sklearn.linear_model import Ridge
lr2=Ridge().fit(X_train,y_train)
lr2.score(X_test,y_test)
```

```python
#As we can see there is no change in R2 score even after the L1 and L2 Regularization , so ther is no overfitting in Linear Regression
```

```python
from sklearn.cluster import KMeans #KMeans is a popular clustering algorithm used for unsupervised machine learning tasks
```

```python
#KMeans partitions n observations into k clusters, where each observation belongs to the cluster with the nearest mean (cluster center).
#The algorithm works iteratively to assign each data point to one of k centroids, forming clusters. 
#The centroids are then updated to be the mean of all the data points assigned to that centroid.
```

```python
kmeans_model=KMeans(n_clusters=200)#Here,the algorithm will attempt to partition the data into 200 clusters.
```

```python
param['clusters']=kmeans_model.fit_predict(param)
#Here, KMeans clustering model is used to predict cluster assignments for a dataset, and then we are storing these cluster assignments in a variable
```

```python
center= kmeans_model.cluster_centers_ # this retrieves the coordinates of the cluster centers that the KMeans algorithm has computed after fitting the model to the data
```

```python
dataset_center = pd.DataFrame(center, columns=['TLR', 'RPC', 'GO', 'OI', 'Perception', 'Score'])
#Here 'pd.DataFrame' function is used to create a DataFrame from the center array
```

```python
dataset_center.sort_values(by='Score', ascending=False) 
#this will sort the dataset_center DataFrame by the 'Score' column in descending order
```

```python
def get_estimated_score(rank): # this function takes a rank as input and retrieves the corresponding estimated score from the 'Rank_score' column in the DataFrame 
    estimated_score = Rank_score.iloc[rank - 1][0]#retrieve the estimated score corresponding to the given rank
    print('Estimated Score:', estimated_score) #prints  the estimated 'Score' correspond to the 'Rank'
    return estimated_score

testing_rank = 20 # Example usage of the function
estimated_score = get_estimated_score(testing_rank)
```

```python
def find_closest_parameters(dataset_center, estimated_score):
    absolute_diff = (dataset_center['Score'] - estimated_score).abs()
    closest_index = absolute_diff.idxmin()
    estimated_parameter_values = dataset_center.loc[[closest_index]]
    return estimated_parameter_values


estimated_parameter_values = find_closest_parameters(dataset_center, estimated_score)

print('Estimated Parameter Values:')
print(estimated_parameter_values)


#find_closest_parameters function you provided takes the dataset_center DataFrame and an estimated_score as inputs.
#It calculates the absolute differences between the 'Score' column in the dataset_center DataFrame and the estimated_score,--> 
#--> finds the index corresponding to the minimum absolute difference, -->
#and then returns the parameter values from the dataset_center DataFrame for that index
```
